# Drop duplicate rows/columns

````python
# Step 1: remove exact duplicates
listings = listings.loc[:, ~listings.T.duplicated()]

# Step 2: remove .1 columns (extra cleanup)
listings = listings.loc[:, ~listings.columns.str.endswith('.1')]

listings.info()
````

# 1. Fixing columns that has dates
```
3   CloseDate                     4995 non-null   object 
48  TaxYear                       27 non-null     float64
55  PurchaseContractDate          9248 non-null   object 
56  ListingContractDate           22058 non-null  object 
```

````python
# changing it to a datetime64[ns]
date_cols = ['CloseDate', 'PurchaseContractDate', 'ListingContractDate']

for col in date_cols:
    listings[col] = pd.to_datetime(listings[col], errors='coerce')

listings[date_cols].dtypes
````
